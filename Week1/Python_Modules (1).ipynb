# Matplotlib Implementation

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 3, 4])
ypoints = np.array([2, 4, 6, 8])

# Plot these points without drawing a line
plt.scatter(xpoints, ypoints, color='black') 
plt.show()

# Plotting with marker: Plot these points with a star marker
plt.scatter(xpoints, ypoints, marker='*', color='black')  
plt.show()

# Using fmt format: Add circular marker, red color, and dashed line
plt.plot(xpoints, ypoints, 'ro--')  
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Plot with Circular Marker, Red Color, and Dashed Line')
plt.show()

# Create a scatter plot for xpoints and ypoints
plt.scatter(xpoints, ypoints)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot of Points')
plt.show()

# Set color to the scatter plot. Blue, Green, Red, and Yellow color for each point respectively
colors = ['blue', 'green', 'red', 'yellow']
plt.scatter(xpoints, ypoints, color=colors)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot with Different Colors')
plt.show()

```
